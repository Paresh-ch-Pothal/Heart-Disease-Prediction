```python
import numpy as np
import pandas as pd

from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import MinMaxScaler,LabelEncoder
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv("heart_disease.csv")
df
```

```python
label_encoder = LabelEncoder()
df['Heart_ stroke'] = label_encoder.fit_transform(df['Heart_ stroke'])
```

```python
df=df.drop(columns=["education","prevalentStroke","prevalentHyp"])
df
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(df.drop(columns=["Heart_ stroke"],axis=1),df["Heart_ stroke"]
                                           ,test_size=0.2,random_state=0)
```

```python
df.isnull().sum()
```

```python
si_cpd=SimpleImputer()
si_bpm=SimpleImputer()
si_tc=SimpleImputer()
si_bmi=SimpleImputer()
si_hr=SimpleImputer()
si_gl=SimpleImputer()
```

```python
xtrain
```

```python
xtrain_cpd=si_cpd.fit_transform(xtrain[["cigsPerDay"]])
xtrain_bpm=si_bpm.fit_transform(xtrain[["BPMeds"]])
xtrain_tc=si_tc.fit_transform(xtrain[["totChol"]])
xtrain_bmi=si_bmi.fit_transform(xtrain[["BMI"]])
xtrain_hr=si_hr.fit_transform(xtrain[["heartRate"]])
xtrain_gl=si_gl.fit_transform(xtrain[["glucose"]])

xtest_cpd=si_cpd.fit_transform(xtest[["cigsPerDay"]])
xtest_bpm=si_bpm.fit_transform(xtest[["BPMeds"]])
xtest_tc=si_tc.fit_transform(xtest[["totChol"]])
xtest_bmi=si_bmi.fit_transform(xtest[["BMI"]])
xtest_hr=si_hr.fit_transform(xtest[["heartRate"]])
xtest_gl=si_gl.fit_transform(xtest[["glucose"]])
```

```python
ohe_gender=OneHotEncoder(sparse_output=False)
xtrain_gender=ohe_gender.fit_transform(xtrain[["Gender"]])

xtest_gender=ohe_gender.fit_transform(xtest[["Gender"]])
```

```python
xtrain
```

```python
xtrain_rem=xtrain.drop(columns=["Gender","cigsPerDay","BPMeds","totChol","BMI","heartRate","glucose"])
xtest_rem=xtest.drop(columns=["Gender","cigsPerDay","BPMeds","totChol","BMI","heartRate","glucose"])
```

```python
xtrain_rem
```

```python
xtrain_trans = np.concatenate((xtrain_rem,xtrain_cpd,xtrain_bpm,xtrain_tc,xtrain_bmi,xtrain_hr,xtrain_gl,xtrain_gender),axis=1)
xtest_trans = np.concatenate((xtest_rem,xtest_cpd,xtest_bpm,xtest_tc,xtest_bmi,xtest_hr,xtest_gl,xtest_gender),axis=1)
```

```python
xtrain_trans.shape
```

```python
clf=DecisionTreeClassifier()
clf.fit(xtrain_trans,ytrain)
```

```python
ypred=clf.predict(xtest_trans)
```

```python
ypred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,ypred)
```
